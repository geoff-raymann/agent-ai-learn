# Groq API Interaction
This notebook demonstrates how to interact with the Groq API using Llama models through a simple question-answer interface.

## Setup: Import Libraries and Initialize Groq Client

```python
import os
from dotenv import load_dotenv
from groq import Groq
import textwrap
import ipywidgets as widgets
from IPython.display import display, clear_output
```

```python
# Load environment variables from .env file
load_dotenv()

# Initialize Groq client with your API key
client = Groq(api_key=os.getenv("GROQ_API_KEY"))

# Set the model and parameters
model_name = "llama-3.3-70b-versatile"
system_prompt = "You are a senior telecom strategy consultant. Answer the question as concisely as possible, and provide actionable insights. If you don't know the answer, say you don't know. Respond in French. This is a question and answer session, so be concise and to the point. Do not provide any additional information or context beyond what is necessary to answer the question."
```

## Generate Content from User Input

## Ask a Question
Edit the `question` variable in the cell below, then run both cells.

```python
# ✏️ Edit your question here, then run this cell
question = "What are the top 3 risks in a telecom project?"
```

```python
try:
    response = client.chat.completions.create(
        model=model_name,
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": question}
        ],
        temperature=0.7,
        max_tokens=300
    )
    answer_text = response.choices[0].message.content
    wrapped_answer = textwrap.fill(answer_text, width=80)
    print("=" * 60)
    print("Answer")
    print("=" * 60)
    print(wrapped_answer)
    print("=" * 60)
except Exception as e:
    print(f"\nError: {type(e).__name__}")
    print(f"Message: {str(e)}")
```
